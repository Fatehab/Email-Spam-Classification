```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, plot_confusion_matrix
```

```python
df = pd.read_csv("/content/drive/MyDrive/emails.csv")
df.head(20)
```

# Splitting Data

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

# Naive Bayes

```python
mnb = MultinomialNB(alpha=1.9)
mnb.fit(train_x,train_y)
y_pred1 = mnb.predict(test_x)
mnb_acc = round(accuracy_score(y_pred1,test_y)*100,2)
print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,test_y))
```

```python
plot_confusion_matrix(mnb,test_x,test_y)
```

# Support Vector Machines

```python
svc = SVC(C=1.0,kernel='rbf',gamma='auto')
svc.fit(train_x,train_y)
y_pred2 = svc.predict(test_x)
svc_acc = round(accuracy_score(y_pred2,test_y)*100,2)
print("Accuracy Score for SVC : ", accuracy_score(y_pred2,test_y))
```

```python
plot_confusion_matrix(svc,test_x,test_y)
```

# Random Forests (Bagging)

```python
rfc = RandomForestClassifier(n_estimators=100,criterion='gini')
rfc.fit(train_x,train_y)
y_pred3 = rfc.predict(test_x)
rfc_acc = round(accuracy_score(y_pred3,test_y)*100,2)
print("Accuracy Score of Random Forest Classifier : ", accuracy_score(y_pred3,test_y))
```

```python
plot_confusion_matrix(rfc,test_x,test_y)
```

As expected, Random Forest Classifier performs the best among the three. Decision tree classifiers are excellent classifiers. Random forest is a popular ensemble model that uses a forest of decision trees. So, obviously, combibining the accuracy of 100 trees (as n_estimators=100 here), will create a powerful model.

# Logistic Regression

```python
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(train_x,train_y)
y_pred4 = lr_model.predict(test_x)
lr_acc = round(accuracy_score(y_pred4,test_y)*100,2)
print("Accuracy Score for SVC : ", accuracy_score(y_pred4,test_y))
```

```python
plot_confusion_matrix(lr_model,test_x,test_y)
```

# XGBoost

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
clf.fit(train_x,train_y) 
y_pred5 = clf.predict(test_x)
clf_acc = round(accuracy_score(y_pred5,test_y)*100,2)
print("Accuracy Score for SVC : ", accuracy_score(y_pred5,test_y))
```

```python
plot_confusion_matrix(clf,test_x,test_y)
```
